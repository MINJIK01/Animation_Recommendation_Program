```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('character1.csv', encoding='utf-8')
```

```python
allData = pd.read_csv('최종데이터.csv', encoding='utf-8')
```

```python
tag = []
```

```python
tag = data.tag1
```

```python
tag
```

```python
for i in range(len(data.tag1)):
    tag[i] = tag[i].split('/')
```

```python
for i in range(len(data.tag1)):    
    if pd.isna(data.tag2[i]) == False :
        tag[i].append(data.tag2[i])
for i in range(len(data.tag1)):    
    if pd.isna(data.tag3[i]) == False :
        tag[i].append(data.tag3[i])
for i in range(len(data.tag1)):    
    if pd.isna(data.tag4[i]) == False :
        tag[i].append(data.tag4[i])
for i in range(len(data.tag1)):    
    if pd.isna(data.tag5[i]) == False :
        tag[i].append(data.tag5[i])
for i in range(len(data.tag1)):    
    if pd.isna(data.tag6[i]) == False :
        tag[i].append(data.tag6[i])
```

```python
data['tag'] = tag
```

```python
data.drop(['tag1', 'tag2', 'tag3', 'tag4', 'tag5', 'tag6'], axis = 1, inplace = True)
```

```python
allData.drop(['줄거리', '제작정보', '태그', '태그2', '성비', '평점', '평점수', '장르'], axis = 1, inplace = True)
```

```python
result = pd.merge(data, allData, on = '제목', how = 'right')
```

```python
result
```

```python
tag_2 = result[['tag']]
tag_2
```

```python
list_tag_2 = tag_2.values.tolist()
```

```python
s_tag_2 = [0]*2120

for cnt in range(2120):
    s_tag_2[cnt] = ''.join(str(list_tag_2[cnt]))

for i in range(2120):
        s_tag_2[i] = s_tag_2[i].replace('[','')
        s_tag_2[i] = s_tag_2[i].replace("]",'')
        s_tag_2[i] = s_tag_2[i].replace("'",'')
        s_tag_2[i] = s_tag_2[i].replace("'",'')
s_tag_2
```

```python
result['tag'] = s_tag_2
```

```python
result
```

```python
result.to_csv('all_ani_character2.csv',encoding='utf-8-sig')
```
