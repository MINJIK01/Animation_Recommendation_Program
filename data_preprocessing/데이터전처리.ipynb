```python
########제작 컬럼에 '각본'이 들어간 행 정보 저장 및 그림 컬럼에 옮겨서 저장.

jeajak_gakbon = []*2120

jeajak_gakbon = final['제작'][gakbon[0]]
```

```python
########제작 컬럼에 '출시'가 들어간 행 정보 저장 및 그림 컬럼에 옮겨서 저장.
jeajak_chulsi = []*2120
jeajak_chulsi = final['제작'][chulsi[0]]
```

```python
def cleaning(x) :
  if str(x).find("제작 : ") != -1 :
    return str(x).replace("제작 : ","") 
  return str(x)
def cleaning2(x) :
  if str(x).find("그림 : ") != -1 :
    return str(x).replace("그림 : ","") 
  return str(x)
def cleaning3(x) :
  if str(x).find("출시 : ") != -1 :
    return str(x).replace("출시 : ","") 
  return str(x)
def cleaning4(x) :
  if str(x).find("각본 : ") != -1 :
    return str(x).replace("각본 : ","") 
  return str(x)
```

```python
final['제작']=final['제작'].apply(cleaning)
final['각본']=final['각본'].apply(cleaning4)
final['그림']=final['그림'].apply(cleaning2)
final['출시']=final['출시'].apply(cleaning3)
final
```

```python
rd_loc = 0 

final_df = pd.DataFrame(columns = ['제목', '장르', '줄거리','평점', '태그'])

for i , row in final.iterrows():
    title = row['제목'] 
    genre = row['장르']
    story = row['줄거리']
    rate = row['평점']
    
    for s in row['태그완성'].split('#'):
        s = str(s).strip() 
        final_df.loc[rd_loc] = [title, genre, story, rate, s] 
        rd_loc = rd_loc + 1
```
